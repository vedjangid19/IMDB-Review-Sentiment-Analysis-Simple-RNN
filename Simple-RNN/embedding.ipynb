```python
from tensorflow.keras.preprocessing.text import one_hot # type: ignore
from tensorflow.keras.utils import pad_sequences # type: ignore
from tensorflow.keras.layers import Embedding # type: ignore
from keras.layers import Embedding
from tensorflow.keras.models import Sequential # type: ignore
import numpy as np
```

```python
### sentences
sent=[
    'the glass of milk',
    'the glass of juice',
    'the cup of tea',
    'I am a good boy',
    'I am a good developer',
    'understand the meaning of words',
    'your videos are good',
]
```

```python
vocab_size = 10000
```

```python
one_hot_repr = [one_hot(words, vocab_size) for words in sent]
one_hot_repr
```

```python
sent_length = 8
```

```python
embedded_docs = pad_sequences(one_hot_repr, padding='post', maxlen=sent_length)
embedded_docs
```

```python
diamention = 10
model = Sequential()
model.add(Embedding(vocab_size, diamention, input_length=sent_length))
model.compile('adam', 'mse')

# Build the model explicitly (optional for debugging)
model.build(input_shape=(None, sent_length))

model.summary()
```

```python
model.predict(embedded_docs)
```
